```python
show=df_app['No-show'] == 'No'
noshow=df_app['No-show'] == 'Yes'
```

# plotting histogram for Age by no show

```python
plt.figure(figsize=(10,6))
plt.hist(df_app[show]['Age'],bins=15,color='red',alpha=0.5,label='SHOW')
plt.hist(df_app[noshow]['Age'],bins=15,color='blue',alpha=0.5,label='NO SHOW')
plt.title("Ages By No Show")
plt.xlabel('Age')
plt.ylabel('Patient num')
plt.legend()
plt.show()
```

# plotting barplot to know that the  age and chronic diseases affect the attendance together

```python
plt.figure(figsize=(16,4))
df_app[show].groupby(['Hipertension','Diabetes'])['Age'].mean().plot(kind='bar',color='pink',label='show')
df_app[noshow].groupby(['Hipertension','Diabetes'])['Age'].mean().plot(kind='bar',color='blue',label='noshow')
plt.title('comparision according to age, chroinc diseases')
plt.xlabel('chronic diseases')
plt.ylabel('mean age')
plt.legend()
plt.show()
```

```python
df_app[show].groupby(['Hipertension','Diabetes'])['Age'].mean()
```

# plotting the pie chart for comparing between attendance by gender.

```python
df_app['Gender'][show].value_counts(normalize=True).plot(kind='pie',label="show")
plt.title("Comparing the attendance by gender for show column")
plt.xlabel("Gender")
plt.ylabel("Patients Number")
plt.legend()
plt.show()
```

## Plotting hist graph to know that  reciving sms affect the attend.

```python
df_app['SMS_received'][show].hist(alpha = 0.5, bins = 12, color ='red',label ='show')
df_app['SMS_received'][noshow].hist(alpha = 0.5, bins = 12, color ='blue',label ='noshow')
plt.title("Comparing accordinng to sms receiving")
plt.xlabel("SMS")
plt.ylabel("Patient num")
plt.legend()
plt.show()
```

# Number of showing patients without receiving sms is greater than showing patients with receiving sms, which mean that we need to revisit our sms campaign

```python
#does neighbourhood affect the attendance
plt.figure(figsize=(14,8))
df_app['Neighbourhood'][show].value_counts().plot(kind='bar', color = 'blue', label = 'show')
df_app['Neighbourhood'][noshow].value_counts().plot(kind='bar', color = 'red', label = 'noShow')

plt.title('comparison according to Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Patient Num')
plt.legend()
plt.show()
```

# This means neighbourhood has a great affect on the attendance

```python
plt.figure(figsize=(14,8))
df_app[show].groupby('Neighbourhood').SMS_received.mean().plot(kind = 'bar', color = 'blue', label = 'show')
df_app[noshow].groupby('Neighbourhood').SMS_received.mean().plot(kind = 'bar', color = 'red', label = 'noShow')

plt.title('Comparison according to Neighbourhood, SMS_receiving')
plt.xlabel('Neighbourhood')
plt.ylabel('patients num')
plt.legend()
plt.show()
```

# SMS has response in only 5 neighbourhoods

```python
plt.figure(figsize=(14,8))
df_app[show].groupby('Neighbourhood').Age.mean().plot(kind = 'bar', color = 'blue', label = 'show')
df_app[noshow].groupby('Neighbourhood').Age.mean().plot(kind = 'bar', color = 'red', label = 'noShow')

plt.title('Comparison according to Neighbourhood, Age')
plt.xlabel('Neighbourhood')
plt.ylabel('mean Age')
plt.legend()
plt.show
```
