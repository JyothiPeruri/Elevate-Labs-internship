In machine learning, we usually deal with datasets that contain multiple labels in one or more than one columns. These labels can be in the form of words or numbers. To make the data understandable or in human-readable form, the training data is often labelled in words.


```python
emp_attri_predi['Attrition'] = emp_attri_predi['Attrition'].replace({'No':0,'Yes':1})
```

```python
#encode binary variables
emp_attri_predi['OverTime'] = emp_attri_predi['OverTime'].map({'No':0,'Yes':1})
emp_attri_predi['Gender'] = emp_attri_predi['Gender'].map({'Male':0,'Female':1})
emp_attri_predi['Over18'] = emp_attri_predi['Over18'].map({'Y':1,'No':0})

```

```python
# encode categorical columns which are ordinal, use LabelEncoding
# apply label encoder to df_categorical
from sklearn.preprocessing import LabelEncoder
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders = {}
for column in encoding_cols:
    label_encoders[column] = LabelEncoder()
    emp_attri_predi[column] = label_encoders[column].fit_transform(emp_attr_predi[column])



```

# Machine Learning :splitting the data into training and testing sample


```python
x=emp_attri_predi.drop(['Attrition','Over18'],axis=1)
y=emp_attri_predi['Attrition'].values
```

# Resampling

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus=RandomOverSampler(random_state = 42)
x_over, y_over = rus.fit_resample(x,y)
print(Counter(y_over))

 
```

```python
# split the data into trainung and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x_over, y_over, test_size=0.2, random_state =42)
```

```python
#sanity check for the sampled data
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Logistic Regression in machine learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
logreg= LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
prediction=logreg.predict(X_test)
cnf_matrix=confusion_matrix(y_test,prediction)
print("Accuracy Score-",accuracy_score(y_test,prediction))
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot = True, cmap = 'Blues', fmt = 'd')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

ax2 = fig.add_subplot(1,2,2)
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  prediction)
auc = roc_auc_score(y_test, prediction)
ax2 = plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
